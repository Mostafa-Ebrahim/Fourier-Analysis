## Spectrogram Example

```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
dt = 0.001
t = np.arange(0,2,dt)
f0 = 50
f1 = 250
t1 = 2
```

```python
x = np.cos(2*np.pi*t*(f0 + (f1-f0)*np.power(t,2)/(3*t1**2)))
```

```python
plt.specgram(x, NFFT=128, Fs=1/dt, noverlap=120,cmap='jet')
plt.colorbar()
plt.show()
```

```python
sd.play(2*x, 1/dt)
```
