## Spectrogram for Beethoven sonnet

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import sounddevice as sd
import librosa
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
y, sr = librosa.load('./beethoven.mp3')
```

```python
plt.specgram(y[0:1000000], NFFT=5000, Fs=sr, noverlap=400,cmap='jet')
plt.colorbar()
plt.show()
```

```python
sd.play(y[:500000], sr)
```
