# Image compression with FFT

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [14, 14]
plt.rcParams['font.size'] = 12
```

```python
photo = imread('IMG.jpg')
grey = np.mean(photo, -1)     # convert RGB into Greyscale
plt.figure()
plt.imshow(photo)
plt.axis('off')
```

```python
greyt = np.fft.fft2(grey)
greytsort = np.sort(np.abs(greyt.reshape(-1)))
```

```python
# Zero out all small coefficients and inverse transform
for keep in (0.1, 0.05, 0.01, 0.002):
    thresh = greytsort[int(np.floor((1-keep)*len(greytsort)))]
    ind = np.abs(greyt)>thresh          # Find small indices
    Atlow = greyt * ind                 # Threshold small indices
    Alow = np.fft.ifft2(Atlow).real  # Compressed image
    plt.figure()
    plt.imshow(Alow,cmap='gray')
    plt.axis('off')
    plt.title('Compressed image: keep = ' + str(keep))
```
